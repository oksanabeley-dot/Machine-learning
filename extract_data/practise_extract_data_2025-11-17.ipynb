```python
# Task 1. Create file using context manager and mode w. Give name task_1.txt. Write random number between 1 and 100

# First part - generate random value
from random import randint # built-in lib, –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü—ñ—ó —Ä–∞–Ω–¥–æ–º–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–Ω—å 

def generate_random_number(start:int = 1, end: int = 100) -> int: # —Ñ—É–Ω–∫—Ü—ñ—è –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü—ñ—ó —Ä–∞–Ω–¥–æ–º–Ω–∏—Ö —á–∏—Å–µ–ª
    return randint(start, end) if start < end else randint(1, 100) # –ø–µ—Ä–µ–≤—ñ—Ä–∫–∞ –Ω–∞ –ø—Ä–∞–≤–∏–ª—å–Ω—É –≥–µ–Ω–µ—Ä–∞—Ü—ñ—é end > start

# Second part - write in file
task_file = 'task_1.txt' # –Ω–∞–∑–≤–∞ —Ñ–∞–π–ª—É –∫—É–¥–∏ —Ç—Ä–µ–±–∞ –∑–±–µ—Ä–µ–≥—Ç–∏
with open(task_file, 'w', encoding = 'utf-8') as file: # context manager + open –¥–ª—è —Ä–æ–±–æ—Ç–∏ –∑ —Ñ–∞–π–ª–æ–º 
    random_value = generate_random_number() # —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ä–∞–Ω–¥–æ–º–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–Ω—è 
    file.write(str(random_value)) # –∑–∞–ø–∏—Å 
    
    
# Self check 
with open(task_file, 'r', encoding = 'utf-8') as file: # context manager + open –¥–ª—è —Ä–æ–±–æ—Ç–∏ –∑ —Ñ–∞–π–ª–æ–º 
    print(file.read())
    
# Second variant solve
with open('task_1.txt', 'w') as file:
    file.write(str(randint(1, 100)))
```

```python
# Task 2. Write in task_1.txt - 10_000 rows with random letters

# Input: file.txt
# Hidden: 1 - use context manager open, file, use writeline
# Hidden: 2 - find solution to generate random letter
# Hidden: 3 - loop step 2 N-times
# Output: file.txt - with data, 10_000 rows with random letters

import random # –¥–ª—è –≤–∏–±–æ—Ä—É —Ä–∞–Ω–¥–æ–º–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–Ω—è
import string # –æ—Ç—Ä–∏–º–∞—Ç–∏ —Ä—è–¥–∫–æ–≤—ñ –∑–Ω–∞—á–µ–Ω–Ω—è 

# Part 1. Variant 1. Generate random value
def generate_random_letter(): # —Ñ—É–Ω–∫—Ü—ñ—è –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü—ñ—ó —Ä–∞–Ω–¥–æ–º–Ω–æ—ó –ª—ñ—Ç–µ—Ä–∏
    return random.choice(string.ascii_letters) # –≥–µ–Ω–µ—Ä–∞—Ü—ñ—è 

# Variant 2. Use ord
def generate_random_letter_v2() -> str:
    min_code, max_code = ord('A'), ord('z') # –±–µ—Ä–µ–º–æ –º—ñ–Ω—ñ–º–∞–ª—å–Ω–µ —Ç–∞ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–µ –∑–Ω–∞—á–µ–Ω–Ω—è –≤ —Ç–∞–±–ª–∏—Ü—ñ –∫–æ–¥—É–≤–∞–Ω–Ω—è
    return randint(min_code, max_code)


# Part 2. Save in file. 
N: int = 10_000 # –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Å–∏–º–≤–æ–ª—ñ–≤ 
file_to_save: str = "task_1.txt"

with open(file_to_save, "a", encoding="utf-8") as f: # –≤—ñ–¥–∫—Ä–∏—Ç—Ç—è —Ñ–∞–π–ª—É
    for _ in range(N): # —Ü–∏–∫–ª–æ–º –ø—Ä–æ—Ö–æ–¥–∏–º–æ –¥–ª—è –∑–∞–ø–∏—Å—É
        letters = generate_random_letter() # –≥–µ–Ω–µ—Ä–∞—Ü—ñ—è 
        f.write("\n" + letters) # –∑–∞–ø–∏—Å
```

```python
# Task 3. Calculate stat letters - display in format A: 100, B:50, etc. **add display top N letters, N enter user

# First variant
file_to_save: str = "task_1.txt"

with open(file_to_save, "r", encoding="utf-8") as f: 
    letter_data = f.readlines() # or read

def save_stat(arr: list[str]) -> dict: # —Ñ—É–Ω–∫—Ü—ñ—è –¥–ª—è –∑–±–µ—Ä–µ–∂–µ–Ω–Ω—è 
    letters_stats = {} # –∑–±–µ—Ä—ñ–≥–∞—î–º–æ –≤ —Å–ª–æ–≤–Ω–∏–∫
    for row in arr: # —Ü–∏–∫–ª–æ–º –ø—Ä–æ—Ö–æ–¥–∏–º–æ –≤—Å—ñ –¥–∞–Ω–Ω—ñ
        l = row.strip() # –æ—á–∏—Å—Ç–∫–∞ –≤—ñ–¥ –∑–∞–π–≤–∏—Ö —Å–∏–º–≤–æ–ª—ñ–≤
        if l not in letters_stats:
            letters_stats[l] = 0
        else:
            letters_stats[l] += 1
        
    return letters_stats
    
# Advanced solution 
def top_letters(d: dict[str:int], n:int = 5) -> list[tuple]:
    if not d: # –ø–µ—Ä–µ–≤—ñ—Ä–∫–∞ –Ω–∞ –ø—É—Å—Ç–∏–π —Ñ–∞–π–ª
        return None
    return sorted(d.items(), key = lambda x: x[1])[::-1][:n]
    
stat_data = save_stat(letter_data)

top_l = top_letters(stat_data)


# Second variant 
from collections import Counter # —ñ–º–ø–æ—Ä—Ç –≥–æ—Ç–æ–≤–æ–≥–æ —Ä—ñ—à–µ–Ω–Ω—è –¥–ª—è —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É 

tast_2_file = 'task_1.txt'
with open(tast_2_file, 'r', encoding = 'utf-8') as file:
    data = file.readlines()
    
data = [i.strip() for i in data if i] # –æ—á–∏—Å—Ç–∫–∞ 

freq = Counter(data)
print(freq)

# Show most feeq
print(freq.most_common(5))
```

```python
# Task 1. Read first 3 rows from amazon.csv, display columns 
import csv # —ñ–º–ø–æ—Ä—Ç –±—ñ–±–ª—ñ–æ—Ç–µ–∫–∏ –¥–ª—è —Ä–æ–±–æ—Ç–∏ —ñ–∑ csv

with open("amazon.csv", "r", encoding="utf-8") as f:
    reader = csv.reader(f)

    header = next(reader) # —á–∏—Ç–∞—î–º–æ –∑–∞–≥–æ–ª–æ–≤–∫–∏
    print("Columns:", header)

    for i, row in enumerate(reader, start=1): # —á–∏—Ç–∞—î–º–æ –ø–µ—Ä—à—ñ —Ä—è–¥–∫–∏
        print(row)
        if i == 3:
            break
```

```python
# Task 2. Calculate image_url and display stats 
import csv


# –∑–º—ñ–Ω–Ω—ñ –¥–ª—è –∑–±–µ—Ä–µ–∂–µ–Ω–Ω—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É
total = with_url = without_url = 0

with open("amazon.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f) # —á–∏—Ç–∞—î–º–æ —è–∫ —Å–ª–æ–≤–Ω–∏–∫ –¥–ª—è –∑—Ä—É—á–Ω–æ–≥–æ —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É 

    for row in reader:
        total += 1
        img = row.get("img_link", "").strip() # –æ—á–∏—Å—Ç–∫–∞ —Ç–∞ –æ—Ç—Ä–∏–º–∞–Ω–Ω—è —Ä—è–¥–∫—É

        if img:
            with_url += 1
        else:
            without_url += 1

 
```

```python
# Task 3. Find product_id where id start B09 or B08 
import csv

result = []

with open("amazon.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)

    for row in reader:
        pid = row.get("product_id", "")

        if pid.startswith("B09") or pid.startswith("B08"): # —Ñ—É–Ω–∫—Ü—ñ—ó —Ä—è–¥–∫—ñ–≤ 
            result.append(row.get("product_id"))

```
