#### Крок 5. KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn = KNeighborsClassifier() # n_neighbors
knn.# use fit

y_pred = knn.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
scores = []
n_values = range(1, 21, 2)
for k in n_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    acc = accuracy_score(y_test, knn.predict(X_test))
    scores.append(acc)
    print(f"k={k:2d}. accuracy={acc:.3f}")
```

```python
plt.plot(n_values, scores, marker="o")
plt.xlabel("k (кількість сусідів)")
plt.ylabel("Accuracy")
plt.title("Вплив k на якість KNN")
plt.show()
```

Крок 6. DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier

# use DecisionTreeClassifier
tree = DecisionTreeClassifier(random_state=42)
tree.#use fit 

accuracy_score(y_test, tree.predict(X_test))
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(18, 8))
plot_tree(
    tree,
    feature_names=X.columns,
    class_names=data.target_names,
    filled=True,
    max_depth=3
)
plt.title("Decision Tree (перші 3 рівні)")
plt.show()
```

```python
# Fine-tuning
from sklearn.model_selection import GridSearchCV

param_grid = {
    "max_depth": [2, 3, 4, 5, 6],
    "min_samples_leaf": [1, 5, 10, 20]
}

grid = GridSearchCV(
    DecisionTreeClassifier(random_state=42),
    param_grid,
    cv=5,
    scoring="accuracy"
)

grid.fit(X_train, y_train)
grid.best_params_
```
