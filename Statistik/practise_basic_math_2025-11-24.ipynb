**Задача 1** \
Дані — перегляд (хв): \
[3, 4, 5, 6, 7]

Знайти медіану.

```python
arr = [4, 5, 3, 7, 6]

arr.sort() # sorted 

n = len(arr)

if n % 2 == 1:
    median_ = arr[n // 2]
else:
    median_ = (arr[n // 2 - 1] + arr[n // 2]) / 2

print(median_)

```

**Задача 2** \
Дані: [3, 3, 4, 100] \
Знайти median та mean. Що про них можна сказати?

```python
arr = [3, 3, 4, 100]

# task 1
temp = sorted(arr) # [::-1]

n = len(arr)

if n % 2 == 1:
    median_ = temp[n // 2]
else:
    median_ = (temp[n // 2 - 1] + temp[n // 2]) / 2

# 100, 1, 5, 100, 150, 10

avg_arr = sum(arr) / n
print(median_, avg_arr)

```

**Задача 1** \
Дані: [10, 12, 12, 15, 20]\
Знайти моду.

```python
#  1 variant
arr = [10, 12, 12, 15, 20]
ans = {}
for num in arr:
    if num in ans:
        ans[num] += 1
    else:
        ans[num] = 1
max_count = 0
max_value = 0
for key, value in ans.items():
    if value > max_count:
        max_count = value
        max_value = key
print(max_value)

# 2 variant
from collections import Counter
ans = Counter(arr)
print(ans.most_common(1))

```

# Задача 1
Дані: [2, 4, 6, 8]

Потрібно:
- знайти середнє
- знайти квадрати відхилень
- знайти дисперсію
- інтерпретувати результат


```python
arr_ = [2, 4, 8, 6]

avg_ = sum(arr_) / len(arr_)
step_2 = [num - avg_ for num in arr_]
step_3 = [i ** 2 for i in step_2]
var_ = sum(step_3) / len(arr_)
std_ = var_ * (0.5)
cv_ = std_ / avg_
print(cv_)
print(std_)
print(var_)
```

Задача 2 — два набори з однаковим середнім, але різним розкидом \
A: [4, 5, 6] \
B: [1, 5, 9] 

Середні однакові.  
Який набір стабільніший?

```python
arr_1 = [4, 5, 6]
arr_2 = [1, 5, 9]

avg_1 = sum(arr_1) / len(arr_1)
step_2 = [num - avg_1 for num in arr_1]
step_3 = [i ** 2 for i in step_2]
var_1 = sum(step_3) / len(arr_1)
std_1 = var_1 * (0.5)
cv_1 = std_1 / avg_1 
print(cv_1)

avg_2 = sum(arr_2) / len(arr_2)
step_2 = [num - avg_2 for num in arr_2]
step_3 = [i ** 2 for i in step_2]
var_2 = sum(step_3) / len(arr_2)
std_2 = var_2 * (0.5)
cv_2 = std_2 / avg_2
print(cv_2)


```
