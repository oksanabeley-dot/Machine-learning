### Основні операції з векторами

#### Додавання векторів

**Формула:** (a1​,a2​,...,an​)+(b1​,b2​,...,bn​)=(a1​+b1​,a2​+b2​,...,an​+bn​) \
**Обмеження:** Додавати можна тільки вектори однакової розмірності.
Для чого:
> рух \
> агрегація \
> matching features



**Властивості**:
> **Комутативність**: v + w = w + v. Порядок не важливий \
> **Ассоціативність**: (u + v) + w= u + ( v + w).  Групувати можна як гручно \
> **0-вектор**: v + 0 = v \
> **minus вектор**: v + (−v) = 0. Відмінити дію. 

---
#### Віднімання векторів
**Формула**: a−b=(a1​−b1​,a2​−b2​,...,an​−bn​)

Приклад: (4,2)−(1,5)=(3,−3)

**Застосування:**
* вектор напрямку: direction=B−A
* обчислення відстані: |B−A|
* знаходження різниці ознак
* помилка моделі: error=ytrue​−ypred​

---

#### **Множення вектора на число**

**Формула:** k(a1​,a2​,...,an​)=(ka1​,ka2​,...,kan​)

Це **масштабування** вектора:

* ( k > 1 ) → збільшуємо довжину
* ( 0 < k < 1 ) → зменшуємо
* ( k < 0 ) → міняємо напрям на протилежний

Приклад

[
2(3, -1) = (6, -2)
]

**Застосування:**

* регулювання швидкості/сил у фізиці
* зміна інтенсивності ознак
* масштабування градієнтів
* нормалізація аудіо / зображень

---

#### **Довжина (норма) вектора**

Довжина = “наскільки великий” вектор.

**Застосування**
* відстань між точками
* нормалізація ознак
* нормування embeddings
* довжина градієнта
* швидкість/сила у симуляціях

---

#### **Нормалізація вектора**

Беремо напрямок вектора, але “обнуляємо” масштаб.
Отримуємо вектор довжиною 1.

**Застосування:**

* cosine similarity
* стабільність моделі
* усунення масштабних перекосів у даних
* напрямок руху об’єкта

---

### **Скалярний добуток (dot product)**

**Формула:** v⋅w = x1​y1​+ x2​y2​+ ... + xn​yn​

#####  Геометричний зміст

* антиколінеарність (кут 180°): < 0

* cosine similarity в NLP/embeddings
* фільтри в Convolutional Neural Networks
* обчислення кута між силами/швидкостями
* пошук схожих зображень
* детекція аномалій

````python
# Знайти, хто найближчий до користувача u1
import numpy as np

u1 = np.array([0.2, 0.8, 0.3, 0.1])
users = np.array([
    [0.25, 0.82, 0.28, 0.12],
    [0.9, 0.1, 0.2, 0.05],
    [0.18, 0.75, 0.32, 0.14]
])

def cos_sim(a, b):
    return np.dot(a,b) / (np.linalg.norm(a) * np.linalg.norm(b))

for i,u in enumerate(users, start=2):
    print(f"cos(u1, u{i}) =", cos_sim(u1, u))
````

# Основні операції з векторами 

| Операція | Застосування | Інтерпретація | Приклад | Обмеження / Правила |
|---------|--------------|---------------|----------|----------------------|
| **Додавання**  \((v + w)\) | Об'єднання ознак, сумарне переміщення, агрегація градієнтів | Кожна координата додає свій “вклад” | **Формула:**  \((3,1)+(2,-4)=(5,-3)\) <br> **Python:** `v+w` | ▪ Вектори **мають бути однакової розмірності** <br> ▪ Комутативність: \(v+w=w+v\) <br> ▪ Асоціативність: \((u+v)+w=u+(v+w)\) |
| **Віднімання** \((v - w)\) | Напрямок між точками, різниця ознак, помилка моделі | Показує, куди та наскільки зміститися від w до v | **Формула:** \((4,2)-(1,5)=(3,-3)\) <br> **Python:** `v-w` | ▪ Розмірності **мають збігатися** <br> ▪ Не комутативне: \(v-w \ne w-v\) |
| **Множення на число** \((k \cdot v)\) | Масштабування ознак, швидкості, сил; контроль градієнтів | Змінює довжину; знак k змінює напрям | **Формула:** \(2(3,-1)=(6,-2)\) <br> **Python:** `2*v` | ▪ Число — скаляр <br> ▪ Довжина масштабується: \(\|kv\|=|k|\|v\|\) |
| **Норма (довжина)** \(\|v\|\) | Відстань, масштаб ознак, довжина градієнтів | “Наскільки великий” вектор | **Формула:** \(\|(3,4)\|=5\) <br> **Python:** `np.linalg.norm(v)` | ▪ Завжди ≥ 0 <br> ▪ \(\|v\|=0\) тільки для нульового вектора <br> ▪ Норма — узагальнений модуль |
| **Нормалізація** \(\hat{v}=v/\|v\|\) | Cos-similarity, стабільність моделей, однакові масштаби | Довжина стає 1, напрямок зберігається | **Формула:** \((3,4)/5=(0.6,0.8)\) <br> **Python:** `v/np.linalg.norm(v)` | ▪ **Не можна ділити нульовий вектор** <br> ▪ Результат має норму = 1 |
| **Скалярний добуток** \(v\cdot w\) | CNN-фільтри, embeddings, вирівняність ознак | Вимірює, наскільки вектори “дивляться” в один бік | **Формула:** \((3,4)\cdot(1,2)=11\) <br> **Python:** `np.dot(v,w)` | ▪ Розмірність має збігатися <br> ▪ Чутливий до масштабу (довгих векторів) <br> ▪ Результат не нормований |
| **Косинусна схожість** \(\frac{v\cdot w}{\|v\|\|w\|}\) | NLP, CLIP, face-recognition, рекомендації | Чиста схожість за напрямком (кут) | **Формула:** ≈ 0.984 для (3,4) і (1,2) <br> **Python:** `dot/(norm_v*norm_w)` | ▪ Обидві норми мають бути ≠ 0 <br> ▪ Ігнорує масштаб (тільки напрямок) <br> ▪ У великих просторах значення “скупчуються” до 0–1 |
| **Матриця × вектор** \(A v\) | Лінійні шари в ML, PCA, трансформації | Поворот, масштабування, проєкція | **Формула:** \(\begin{pmatrix}1&2\\0&1\end{pmatrix}(3,1)=(5,1)\) <br> **Python:** `A @ v` | ▪ Число стовпців матриці = розмірності вектора <br> ▪ Лінійність: \(A(v+w)=Av+Aw\) |

### ЩО ТАКЕ МАТРИЦЯ

**Матриця** — це прямокутна таблиця чисел, що складається з рядків і стовпців.

**Матриця** — це функція, яка перетворює один вектор на інший. Вектор → матриця → новий вектор.

#### **Cтруктура матриці**
Матриця складається з: \
    - **елементів** \
    - **рядків** \
    - **стовпців** \
    - **розмірності (shape)**

Приклад:
A = [[1, 2, 3],
     [4, 5, 6]]

елементи: 1, 2, 3, 4, 5, 6, 2 рядки, 3 стовпці, shape = (2, 3)

#### Види матриць
> Квадратна матриця (n × n). Однакова кількість рядків і стовпців \
> Прямокутна матриця (m × n). 3×1, 2×3 тощо \
> Діагональна матриця. Всі позакутові елементи = 0 \
> Одинична матриця. Перетворення, яке не змінює вектор \
> Нульова матриця. Всі елементи = 0 \
> Трикутна матриця. Верхня чи нижня

#### **Властивості**
Розмір (shape)

Матриця має розмір m × n, де:
- m — число рядків
- n — число стовпців

Нижче — **повний, чіткий і практичний блок “ВЛАСТИВОСТІ МАТРИЦЬ”**:
теорія → інтуїтивне пояснення → приклади → де реально застосовується в ML / CV / Data Science / аналітиці.

Зроблено у форматі, який можна давати студентам **з нуля**, але який зрозуміє і інженер.

---


#### **Комутативність додавання матриць**

**Властивість:** A + B = B + A

Ми складаємо матриці **покомпонентно**. Порядок не має значення.



**Де використовується:**

* об’єднання фільтрів у CNN
* додавання вагових матриць
* сумарні змішування трансформацій
* аналітика: робота з ковариаційними матрицями

---

#### **Асоціативність додавання**

**Формула**: (A+B)+C = A+(B+C)

Компонентне додавання аналогічне додаванню чисел.

**Практичне застосування**

Можна групувати операції для прискорення навчання (batch-based updates).

---

#### **Закон розподілу для множення числа**

[
k(A + B) = kA + kB
]


Скаляр “розтягує” кожен елемент матриці однаково.

* регуляризація (λ·W)
* масштабування градієнтів
* rescaling ковариацій

---

#### **Асоціативність множення матриць**

[A(BC) = (AB)C]

Бо дозволяє:

* оптимізувати порядок множення
* прискорювати обчислення в нейромережах
* комбінувати кілька лінійних перетворень в одне


Приклад:
У нейронці шар = Wx + b.
Для батчу: [W(XB) = (WX)B]

GPU завжди вибирає оптимальний порядок.

---

#### **Не комутативність множення**

[AB \neq BA]


Множення матриць — це **послідовність перетворень**.
Порядок важливий: виконати поворот → масштабування ≠ масштабування → поворот.


**Де використовується**

* 3D графіка (трансформації: rotate → scale → translate)
* порядкові лінійні операції в ML (слої, проєкції)
* фільтри → нормалізація → активації

---

#### **Розподільний закон для множення**

[A(B + C) = AB + AC] = [(A + B)C = AC + BC]

Практичне застосування:
* оптимізація матричних обчислень
* комп’ютерна графіка: об'єднання трансформацій
* backpropagation: розкладання градієнтів

---

#### **Існування одиничної матриці**

[AI = IA = A]


[
I=\begin{pmatrix}1&0\0&1\end{pmatrix}
]

I — “нічого не робить”. Це аналог числа 1.

**Де використовується**

* ініціалізація параметрів
* збереження форми даних
* перевірка властивостей градієнтів

---

#### **Обернена матриця**

[
AA^{-1} = A^{-1}A = I
]

Обернена матриця “скасовує” дію A.

Коли існує?

Тільки якщо матриця **невироджена** (det(A) ≠ 0).

**Практичне застосування:**

* розв’язання систем рівнянь
* whitening даних
* лінійні перетворення назад
* контроль стабільності моделей

#### **Ранг матриці**
Кількість **лінійно незалежних рядків/стовпців**.


---

#### **Детермінант**
    det(A)


Детермінант показує, **як матриця масштабує об’єм** простору.

* det = 0 → матриця стискає простір у нижчу розмірність → немає оберненої
* det > 1 → розтягує простір
* det < 1 → стискає

```markdown
| Властивість | Інтуїція | Приклад | Практичне застосування |
|-------------|----------|----------|------------------------|
| A+B=B+A | Покомпонентне додавання | (A+B) | Об'єднання параметрів, фільтрів |
| (A+B)+C=A+(B+C) | Групування не змінює результат | (A+B)+C | Оптимізація обчислень |
| k(A+B)=kA+kB | Масштаб однаковий для всіх елементів | kA | Регуляризація, rescaling |
| A(BC)=(AB)C | Послідовні трансформації | (AB)C | GPU-оптимізації, NN-шари |
| AB≠BA | Порядок трансформацій важливий | AB vs BA | 3D графіка, пайплайни ML |
| A(B+C)=AB+AC | Розподіл операцій | AB+AC | Backpropagation |
| AI=IA=A | I нічого не змінює | A·I=A | Збереження форми даних |
| AA⁻¹=I | “Скасування” дії | А⁻¹ | Whitening, інверсії |
| Aᵀ | Перекидання рядків і стовпців | Aᵀ | Covariance, XᵀX |
| A=Aᵀ | Симетрія | Cov матриця | PCA, Graph ML |
| rank(A) | Скільки незалежних ознак | rank | Multicollinearity |
| det(A) | Як трансформація змінює об’єм | det | Gaussian, invertibility |
```




````python
# ТРАНСПОНУВАННЯ - зміна рядків на стовбчики
A = [[1, 2],
     [3, 4]]

AT = [
    [A[0][0], A[1][0]],
    [A[0][1], A[1][1]]
]

print(AT)
````

````python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

print(A.T)
````

````python
# множення матриці на вектор 
import numpy as np

A = np.array([[1, 2],
              [3, 4]])
v = np.array([5, 6])

print(A @ v)
````

````python
# Множеннь: Матриця А комбінує рядки, Матриця B комбінує стовпці, і виходить нова матриця.
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print(A @ B)
````

````python
# додавання матриць
import numpy as np

C = np.array(A) + np.array(B)
print(C)
````

````python
# множення на число 
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

print(2 * A)
````
