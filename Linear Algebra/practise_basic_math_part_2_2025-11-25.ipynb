**Task 5.** Нормалізація поведінкових фіч перед кластеризацією


```python
import numpy as np

users = np.array([
    [10, 3, 1],
    [40, 2, 0],
    [20, 5, 2],
    [15, 1, 1],
    [60, 7, 3]
])

# use linalg.norm(ax)
norm_ = np.linalg.norm(users, axis = 1, keepdims=True)
print(norm_)
# users / by previos step
# np.linalg.norm(users, axis=1, keepdims=True) — це обчислення норм (довжин) векторів уздовж рядків матриці users.
# np.linalg.norm() — функція для обчислення норми (за замовчуванням — евклідової, L2).
# axis=1 означає: обчислюємо норму для кожного рядка окремо.
# keepdims=True зберігає вимірність результату, тобто замість форми (n,) ви отримаєте (n, 1).
# Це корисно, коли пізніше потрібно буде виконувати поелементне ділення users / norms.

normalize_users = users / norm_
print(normalize_users)

# use linalg.norm
# users / by previos step
```
