## Probality Theory
> Генеральна сукупність - уся досліджувана вибірковим методом статистична сукупність об'єктів і/або явищ суспільного життя (одиниць добору), що мають загальні якісні ознаки або кількісні змінні.  
> Вибіркова сукупність (вибірка) - представницька частина генеральної сукупності.

> Доповнення події - це подія, яка не відбудеться. Імовірність того, що подія A не відбудеться, позначається P(A'). Імовірність того, що події A і B відбудуться одночасно, - це ймовірність перетину A і B. Імовірність перетину подій A і B позначається P(A ∩ B).

> Intersection P(A ∩ B)
Ймовірність перетину подій A and B
Union P (A∪B)
Ймовірність A or B


**Простори зразків**  

**Експеримент**: будь-який процес або процедура, для якої можливий більш ніж один результат  
**Простір вибірки**: множина всіх можливих результатів експерименту

#### Події
**Події та доповнення**
> **Події**: підмножина простору вибірки  
> **Ймовірність події A**, P(A), визначається як добуток ймовірностей результатів, що містяться у події A   Кажуть, що подія відбувається, якщо відбувається один з результатів, що містяться в події  
> **Доповнення подій**: подія A' - це подія, яка складається з усього, що не міститься у вибірковому просторі S, тобто з усього, що не міститься в події A.
P(A) + P(A') = 1  
> **Елементарна** (або проста) подія: подія, що складається з одного результату


**Закон великих чисел** - це один з основних результатів теорії ймовірностей, який стверджує, що середнє арифметичне великої кількості незалежних випадкових величин майже завжди буде дуже близьким до їх математичного сподівання.

Чим більше спостережень, тим ближче результат до _справжньої_ ймовірності.

#### Дискретні та неперервні випадкові величини

**Дискретні**
Це величини, які: приймають окремі значення

Приклади:
- покупка: 0 або 1;
- клік: так / ні;
- кількість покупок за день: 0, 1, 2, 3…

Конверсія — дискретна випадкова величина

**Неперервні**

Це величини, які: можуть приймати будь-яке значення в діапазоні

Приклади:
- час на сайті;
- чек покупки;
- тривалість перегляду відео.

Саме такі метрики часто мають складні розподіли

## Розподіл імовірностей
> Розподіл імовірностей є одним з основних понять теорії ймовірностей і використовується в основному для вираження закону ймовірності випадкових величин. Для зручності використання, відповідно до різних типів випадкових величин, розподіл імовірностей набуває різних проявів.
>Розподіл імовірностей містить у собі дискретний розподіл імовірностей і безперервний розподіл імовірностей.
    >>Дискретні дані: дані складаються з окремих значень, кожне з яких має відповідну ймовірність.
    >>Неперервні дані: дані охоплюють діапазон, і будь-яке значення в цьому діапазоні може стати результатом події.

Розподіл — це: як часто зустрічаються різні значення випадкової величини

#### Що таке нормальний розподіл

Нормальний розподіл — це розподіл, у якому:
- більшість значень знаходяться біля середнього;
- чим далі від середнього — тим рідше значення;

> Нормальний розподіл - також відомий як Гауссовий розподіл (Гауссовий розподіл), нормальна крива має форму дзвона, два кінці, середня висота є двосторонньо симетричною, а центральне міністерство має найвищу густину ймовірності. У міру віддалення від середнього щільність імовірності зменшується.

- завантаженість агента

Жоден фактор не домінує, але всі трохи впливають.

```python
def coint_flip():
    return random.randint(0, 1)

# list for add result
result_l = []

def monte_carlo(n):
    plt.subplots(figsize = (12,9))
    result = 0
    
    for i in range(n):
        flip_result = coint_flip()
        result = result +  flip_result
        
        prob_var = result / (i + 1)
        #print(prob_var)
        
        result_l.append(prob_var)
        
    plt.axhline(y = 0.5, color = 'r', linestyle = '-')
    plt.xlabel('Iterations')
    plt.ylabel('Probability')
    plt.plot(result_l)
    return result / n
```
