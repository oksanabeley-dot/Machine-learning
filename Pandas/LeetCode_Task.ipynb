196. We need to remove duplicate emails but keep the one with the smallest id.Pandas makes this simple using the drop_duplicates() method.

Approach
Sort the DataFrame by id (so smaller IDs come first).
Drop duplicate emails, keeping the first occurrence.

```python
import pandas as pd
def delete_duplicate_emails(person: pd.DataFrame) -> None:
    person.sort_values(by='id',ascending=True,inplace=True)
    # .sort_values()	Метод сортировки строк таблицы
    # by='id'	Сортировать по столбцу id
    # ascending=True	По возрастанию (от меньшего к большему). Если False — по убыванию
    # inplace=True	Изменять таблицу на месте, без создания новой
    
    person.drop_duplicates(subset='email', keep='first', inplace=True)
    #.drop_duplicates()	Метод, который удаляет дубликаты строк
    # subset='email'	Учитываем только столбец email для поиска дубликатов
    # keep='first'	Оставляем первую встреченную строку (остальные — удаляем)
    # inplace=True	Изменяем DataFrame на месте, не создавая новый
```
