```python
# Векторна операція -  дія над усім стовпцем одразу
df["created_at"] = pd.to_datetime(df["created_at"], errors="coerce")

df["year"] = df["created_at"].dt.year
df["month"] = df["created_at"].dt.month
df["hour"] = df["created_at"].dt.hour
df["weekday"] = df["created_at"].dt.day_name()

```

```python
df["response_time"] = (
    df["first_response_at"] - df["created_at"]
).dt.total_seconds()
```

```python
df[""] = df["title"].str.contains("|", case=False, na=False)
```

```python
df.groupby("country").filter(lambda g: len(g) > 10000)
```

```python
# assign - для створення нових колонок
df = (
    df.assign(
        year=lambda x: x["timestamp"].dt.year,
        month=lambda x: x["timestamp"].dt.month,
        log_duration=lambda x: np.log1p(x["duration_sec"]),
        is_kz=lambda x: x["country"] == "KZ",
    )
)
```
