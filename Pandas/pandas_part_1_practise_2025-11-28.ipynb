---
**Завдання 1:** Оцінити середню вартість діаманта за категоріями чистоти (clarity) \
**Інструмент для вирішення:**  groupby та agg та agg function

```python
df.groupby('clarity').agg({'price': ['mean']})
```

---
**Завдання 2:** Виявити дешеві діаманти \
**Інструменти:** loc, isin, bool mask \

Дешевий це: carat > 1.2, color = "D" або "E", price < 2000

---
**Завдання 3:** Знайди 10 найдорожчих діамантів у кожній категорії кольору.

```python
df.groupby('color')["price"].max().reset_index().sort_values("price", ascending = False).head(10)
```

---
**Завдання 4:** Знайти середню ціну діамантів різних категорій


```python
mean_price_cut = df.groupby("cut")["price"].mean().reset_index()
mean_price_cut.head()
```

---
**Завдання 5:** Порахуйте, на скільки % дорожчі діаманти кольору D порівняно з кольором G

```python
mean_price = df.groupby("color")["price"].mean()
mean_price
price_D = mean_price["D"]
price_G = mean_price["G"]
print((price_D - price_G) / price_G * 100)
```

```python
# приклад корреляції
df.corr(numeric_only=True)["price"].sort_values(ascending=False)
```
