#### Як провести A/B-тест: покроково

Щоб коректно провести A/B-тест, команда має виконати кілька обовʼязкових кроків.

##### Крок 1. Рандомно розділити трафік

Користувачів необхідно **випадково** розподілити на дві (або більше) рівні групи:

* **Контрольна група (A)** — бачить поточну версію
* **Тестова група (B)** — бачить змінену версію

**Важливо:**
> кожен користувач повинен потрапити **тільки в одну групу**
> і бачити **один і той самий варіант протягом усього тесту**

---

#### Крок 2. Показати різні варіанти

Протягом визначеного періоду:

* групі A показується оригінальний варіант
* групі B — змінений варіант

Змінюється **тільки один елемент**, який і перевіряє гіпотезу
(кнопка, текст, логіка, порядок кроків тощо).

---

#### Крок 3. Зібрати дані та виміряти метрики

Для кожної групи збираються дані за заздалегідь визначеними метриками:
* конверсія
* кліки
* продажі
* час
* завершення кроку

**Важливо:**
> метрика має бути визначена **до запуску тесту**

---

#### Крок 4. Порівняти результати

Порівнюється **ефективність варіантів A та B**:
* чи є різниця між групами
* чи є ця різниця статистично значущою
* чи відповідає вона гіпотезі

---

#### Крок 5. Проаналізувати та прийняти рішення

На основі результатів команда вирішує:

* впроваджувати зміну
* не впроваджувати
* або повторити тест / доопрацювати гіпотезу

**A/B-тест не гарантує «покращення», але гарантує **знання**.**

---
#### Підвищення конверсій

**Конверсія** — це частка користувачів, які виконали цільову дію
(покупка, реєстрація, підписка тощо).

---

#### Економія часу та ресурсів

#### Які запитання допомагає вирішити A/B-тестування

За допомогою A/B-тестів бізнес може відповісти на запитання:

---

#### Основні терміни A/B-тестування

##### Гіпотеза (Hypothesis)
Припущення, яке перевіряється тестом.

**Приклад:**

> Якщо додати зображення продукту поруч з описом,
> кількість переглядів товару зросте на 15%.

---

##### Контрольна група (Control Group)

Користувачі, які бачать **оригінальну версію** продукту (Variant A).

---

##### Тестова група (Test Group)

Користувачі, які бачать **змінену версію** продукту (Variant B).

---

##### Метрика (Metric)

Показник, за яким оцінюється результат тесту:

* конверсія
* кліки
* час на сайті
* відсоток відмов

---

##### p-value

Ймовірність того, що отримана різниця — **випадкова**.

**Зазвичай:** p-value < 0.05 -> результат статистично значущий

---

##### Статистична значущість
Ознака того, що різниця між варіантами з високою ймовірністю **реальна**, а не випадкова.

---

##### AA-тестування

Тест без змін, де обидві групи бачать однакову версію.
Використовується для перевірки:

* коректності рандомізації
* системи аналітики

---

##### A/B/n-тестування

Тестування **більше ніж двох варіантів** одночасно.

---

##### Односторонній vs двосторонній тест

* **Односторонній** — перевіряє ефект у заданому напрямку
* **Двосторонній** — перевіряє будь-яку різницю

---

#### Розмір вибірки та тривалість тесту

Навіщо це потрібно:
Недостатня вибірка -> хибні висновки
Занадто короткий тест -> шум замість сигналу

---

#### Основні параметри

* **Baseline conversion rate** — поточна конверсія
* **Expected lift** — очікуваний ефект
* **Significance level (α)** — зазвичай 0.05
* **Statistical power (1−β)** — зазвичай 0.8

---

#### Розрахунок тривалості тесту

**Формула:**

```
Тривалість (дні) = Розмір вибірки / Середній щоденний трафік
```

Завжди округлюємо **в більший бік**.

---

#### Приклад

* Базова конверсія: 10%
* Очікуваний lift: +2%
* α = 0.05
* Power = 0.8
* Трафік: 1000 користувачів/день

Результат:

* 3623 користувачі на групу
* 7246 загалом
* тривалість ≈ **8 днів**

#### Метрики та типи даних в A/B-тестуванні


A/B-тест **не має сенсу без правильної метрики**.
Саме метрика визначає:

* що саме ми оптимізуємо
* як будемо оцінювати результат
* який статистичний метод застосуємо

**Важливо:**

> одну гіпотезу — **одна основна метрика**

---

#### 1. Що таке метрика

**Метрика** — це числовий показник, який відображає результат або зміну в поведінці користувачів.

##### Приклади метрик:

* конверсія в реєстрацію
* CTR кнопки
* середній чек
* час виконання дії
* кількість кліків

---

#### 2. Основні типи метрик в A/B-тестах

#####  Primary metric (основна метрика)

Це **головна метрика**, заради якої проводиться тест.

Вона:
* напряму пов’язана з гіпотезою
* використовується для прийняття рішення

**Приклад:**

> Якщо змінити текст CTA, **конверсія в покупку** зросте

Primary metric → **конверсія в покупку**

---

##### Secondary metrics (додаткові метрики)

Допомагають:

* зрозуміти контекст
* перевірити побічні ефекти

**Приклади:**

* час на сторінці
* кількість переглядів
* bounce rate

Secondary метрики **не використовуються** для головного рішення.

---

#### Guardrail metrics (захисні метрики)

Метрики, які **не повинні погіршитися**.

Використовуються, щоб:

* не зламати продукт
* не нашкодити UX або бізнесу

**Приклади:**

* швидкість завантаження
* кількість помилок
* churn

Навіть якщо primary metric зросла, але guardrail погіршилась — тест **неуспішний**.

---

#### 3. Типи даних у метриках

Тип даних напряму впливає на:

* вибір статистичного тесту
* спосіб аналізу
* перевірку нормальності

---

#### Бінарні дані (Binary)

Метрика має лише **два значення**:

* 0 / 1
* так / ні
* сконвертувався / не сконвертувався

**Приклади:**
* конверсія
* клік
* реєстрація

**Особливості:**
* базуються на біноміальному розподілі
* нормальність **не перевіряється**

##### Типові тести:
* z-test для пропорцій
* χ²
* Fisher’s exact test

---

#### Лічильні дані (Count)

Цілі числа:
* 0, 1, 2, 3, …

**Приклади:**
* кількість кліків
* кількість переглядів
* кількість дій за сесію

**Особливості:**
* часто асиметричні
* багато нулів
* пуассонівський або overdispersed розподіл

##### Типові підходи:

* Poisson / Negative Binomial
* t-test (обережно)
* bootstrap

---

#### Безперервні дані (Continuous)

**Приклади:**
* час на сайті
* середній чек
* latency
* revenue

**Особливості:**

* часто мають довгий хвіст
* можуть бути не нормальними

##### Типові підходи:

* t-test (за умов)
* log-трансформація
* bootstrap
* Mann–Whitney

---

#### Категоріальні дані (Categorical)
Дані без числового порядку.

**Приклади:**
* тип пристрою
* країна
* канал трафіку

Зазвичай:
* не є метриками напряму
* використовуються для **сегментації**

---

#### 4. Рівні вимірювання даних

| Рівень   | Приклад         | Коментар          |
| -------- | --------------- | ----------------- |
| Nominal  | країна, браузер | без порядку       |
| Ordinal  | рейтинг 1–5     | є порядок         |
| Interval | температура     | різниця має сенс  |
| Ratio    | час, гроші      | є абсолютний нуль |

В A/B найчастіше:
* binary
* ratio

#### Тест для конверсії (Binary)
**Коли застосовуємо:**
- метрика = так / ні
- достатньо користувачів

Приклад:
A: 1000 користувачів -> 100 покупок
B: 1000 користувачів -> 130 покупок

**Що робимо:** порівнюємо дві пропорції

**Використовуємо:**
**z-test для пропорцій**

**Інтерпретація:** Якщо p-value < 0.05 → різниця навряд чи випадкова

#### T-test (безперервні дані)
**Коли застосовуємо:**
> середні значення
> немає перекосу
> n достатньо велике

Приклад:

A: середній час = 35 сек
B: середній час = 32 сек

Інтерпретація: Чи могли ми побачити таку різницю випадково?

````python
# p-value — це:
# Ймовірність отримати такий самий або більш “екстремальний” результат, якщо насправді різниці між A і B немає.

# p-value = 0.02 означає 98%, що B кращий
# p-value = 0.04 означає великий ефект
# p-value = 0.06 означає, що ефекту немає

# !!!! p-value НЕ говорить про розмір ефекту !!!!
# !!!! p-value НЕ говорить про бізнес-користь !!!!

#Чому беруть поріг 0.05
# 0.05 означає:
# 5% шансів помилитися
# 95% довіри, що ефект не випадковий
````

````python
# Довірчий інтервал (Confidence Interval)
# Діапазон значень, у якому з високою ймовірністю лежить реальний ефект

# Приклад
# 1000 users, 120 buy

# 1. Calculate p: p = 120 / 1000 = 0.12
# 2. Standard error: SE = √( p · (1 − p) / n )
# SE = √(0.12 · 0.88 / 1000)
# SE ≈ √(0.0001056)
# SE ≈ 0.0103

# 95% trust -> 1.96 
# 3. Build CI
# CI = p ± 1.96 · SE
# CI = 0.12 ± 1.96 · 0.0103
# CI = 0.12 ± 0.020 -> [0.10 ; 0.14]  ->  [10% ; 14%]


````

````python
import math


n = 1000
successes = 120

p = successes / n

se = math.sqrt(p * (1 - p) / n)

z = 1.96

ci_low = p - z * se
ci_high = p + z * se

p, ci_low, ci_high
````

````python
import numpy as np
from scipy import stats # !!!! Need to install scipy 
````

````python
np.random.seed(42)

A = np.random.normal(loc=40, scale=10, size=100)
B = np.random.normal(loc=36, scale=10, size=100)
````

````python
t_stat, p_value = stats.ttest_ind(A, B, equal_var=False)
t_stat, p_value
````

````python
mean_diff = np.mean(B) - np.mean(A)

se = np.sqrt(
    np.var(A, ddof=1)/len(A) +
    np.var(B, ddof=1)/len(B)
)

z = 1.96
ci_low = mean_diff - z * se
ci_high = mean_diff + z * se

mean_diff, ci_low, ci_high
````
