**A/B-тест** — це контрольований експеримент, у якому користувачі випадково розподіляються між варіантами системи, щоб оцінити причинний вплив зміни на цільову метрику.

#### Причинний ефект (Causal effect)
Причинний ефект — це зміна метрики, яка відбулася саме через наше втручання, а не через зовнішні фактори

A/B-тест застосовується тільки тоді, коли нам потрібно відповісти на питання:
> Чи саме ця зміна призвела до результату? \
> Якщо причинний ефект не потрібен — A/B-тест зайвий


Обовʼязкові компоненти A/B-тесту
> Зміна (treatment) \
> Контрольна група \
> Рандомізація \
> Метрика результату \
> Вимірювання після зміни

##### Зміна (Treatment)
Зміна — це контрольоване втручання, яке потенційно впливає на поведінку користувача.

**Приклади змін:**
> інший дизайн сторінки \
> нова ціна \
> інший текст кнопки \
> новий алгоритм


**Якщо зміна:**
> нечітка \
> багатокомпонентна \
> неконтрольована \
> **A/B-тест не має сенсу, бо результат неінтерпретований.**
--- 
##### Контроль над впливом (Control over treatment)
Контроль над впливом — це можливість однозначно визначити, який користувач яку версію отримав.


**Без контролю:**
> користувачі самі обирають \
> групи системно відрізняються \
> причинний ефект недосяжний
--- 
##### Рандомізація (Randomization)
Рандомізація — випадковий розподіл користувачів між групами, який забезпечує їхню еквівалентність до втручання.

**Навіщо вона потрібна**
> усуває приховані фактори \
> дозволяє порівнювати групи напряму \
> робить оцінку ефекту валідною
---
##### Метрика результату (Outcome metric)
Метрика результату — це кількісний показник, який змінюється внаслідок впливу.

**Хорошо метрика:**
> вимірювана
> стабільна
> повʼязана з бізнес-ціллю
---
##### Інтерференція (Interference)
Інтерференція — ситуація, коли результат одного користувачазалежить від того, що отримали інші користувачі.

**При інтерференції:**
> порушується незалежність спостережень \
> стандартний A/B-тест некоректний


##### Коли A/B-тест є доцільним

A/B-тест можна і потрібно проводити, якщо:
> є чітка зміна \
> є контроль і рандомізація \
> є коректна метрика \
> інтерференція мінімальна

##### Коли A/B-тест НЕ доцільний
> немає контролю над впливом \
> дуже малий трафік \
> метрика надто шумна

#### Baseline
Baseline — це поточний рівень метрики до будь-якого втручання, який використовується як точка відліку для майбутнього A/B-тесту.

**Baseline відповідає на питання:**

> З чим ми будемо порівнювати результат тесту і наскільки стабільна поточна ситуація?

**Без baseline:**
> неможливо обрати MDE \
> неможливо порахувати розмір вибірки \
> неможливо інтерпретувати результат

--- 
#### Метрика результату (Outcome metric)
Outcome-метрика — це кількісний показник,який має змінюватися саме через treatment і відображає бізнес-результат.


#### Primary / Secondary / Guardrail
> Primary — метрика, за якою приймається рішення \
> Secondary — пояснює, чому зʼявився ефект \
> Guardrail — метрика, яку не можна погіршити

**Приклад для датасету E-commercy:**
> Primary: Conversion rate (converted) \
> Secondary: Conversion по країнах девайсах \
> Guardrail: refunds

#### Шум (Variance / Variability)
Шум — це природна варіативність метрики, яка не повʼязана з treatment.

**Ознаки:**
> сильні коливання по днях \
> велика різниця між сегментами \
> низька частота подій (low base rate)

**Чим вищий шум:**
> тим важче знайти ефект \
> тим більший ризик помилки II роду \
> тим більша вибірка потрібна

**Гіпотеза** — це чітке, перевірюване твердження про те, як treatment вплине на outcome-метрику.**
> що саме змінюється \
> яку метрику це має змінити \
> у який бік \
> за яких умов ми визнаємо результат

**Без гіпотези:**
> тест неінтерпретований \
> ризики не контрольовані

**Нульова гіпотеза (H₀)**
H₀ — припущення, що між версіями немає відмінностей.

**Альтернативна гіпотеза (H₁)**
H₁ — припущення, що відмінність існує.

---
#### MDE (Minimum Detectable Effect)
**MDE — це найменший ефект, який:**
> має бізнес-сенс
> ми хочемо гарантовано помітити в тесті

**MDE відповідає на питання:**
> Який ефект нас реально влаштує, а менший — не вартий впровадження?
---
#### Помилка I роду (α)
Помилка I роду — імовірність прийняти ефект, якого не існує.

#### Помилка II роду (β)
Помилка II роду — імовірність не помітити реальний ефект.


| Помилка | Реальний сенс                  |
| ------- | ------------------------------ |
| I роду  | Викотили неефективну зміну     |
| II роду | Відмовилися від корисної зміни |

**Типові значення:**
> α = 0.05 — стандарт \
> α = 0.01 — високий ризик \
> α = 0.10 — швидкі експерименти

α — це допустима ймовірність помилки I роду, тобто прийняття ефекту, якого насправді не існує.
**Чим дорожча помилка I роду, тим менше α.**

---
#### Потужність тесту (Power)
Power = 1 − β — імовірність помітити ефект розміру MDE, якщо він реально існує.

**Типові значення:**
> Power = 80%
> Power = 90%

**Чим дорожча помилка II роду,тим більша потужність потрібна.**


#### Stop-criteria
Stop-criteria — це заздалегідь визначена умова завершення тесту.

**Коректні критерії:**
> досягнутий sample size \
> завершений повний бізнес-цикл

#### Sample Ratio Mismatch (SRM)
SRM — це ситуація, коли фактичний розподіл користувачів між групами істотно відрізняється від запланованого.

**Заплановано:**
> 50% — A \
> 50% — B

**Фактично:**
> 58% — A \
> 42% — B


**SRM означає, що:**
> рандомізація порушена \
> групи системно різні
--- 
#### Еквівалентність груп
Еквівалентність груп — це відсутність системних відмінностей між групами до застосування treatment.

> країна \
> пристрій \
> джерело трафіку 
---
#### Незалежність спостережень
Незалежність означає, що кожне спостереження не впливає на інше.

**Типові порушення:**
> один користувач у двох групах \
> рандомізація по сесіях \
> shared devices

---
#### Data loss
Data loss — це ситуація, коли частина подій не записується або записується вибірково.

**Ознаки:**
> різке падіння подій \
> асиметрія між групами \
> зміна логіки збору

| Ситуація                      | Тест          |
| ----------------------------- | ------------- |
| Бінарна метрика, багато даних | χ²            |
| Бінарна, мало даних           | Fisher        |
| Хочеш зрозуміти формулу       | Z-test        |
| Середні, нормальність         | t-test        |
| Довгий хвіст                  | Mann–Whitney  |

| Тест                   | Тип метрики   | Коли застосовується       | Що перевіряє (суть)                | Інтуїція формули                                                                    | Типові α    | Що означає результат                          |
| ---------------------- | ------------- | ------------------------- | ---------------------------------- | ----------------------------------------------------------------------------------------------------- | ----------- | --------------------------------------------- |
| **χ² (Chi-square)**    | Бінарна (0/1) | Великі вибірки, конверсія | Чи залежить результат від групи    | Порівнює: *наскільки фактичні частоти відрізняються від тих, які ми очікували б, якби ефекту не було* | 0.05        | Є / немає статистичної залежності між A і B   |
| **Fisher Exact**       | Бінарна (0/1) | Малі вибірки              | Те саме, що χ², але без наближення | Рахує **точну ймовірність** отримати таку таблицю, якщо H₀ істинна                                    | 0.05        | Є / немає значущої різниці навіть при малих n |
| **Z-test (пропорції)** | Бінарна (0/1) | Великі вибірки            | Чи відрізняються конверсії         | Дивиться, **на скільки стандартних помилок** різниця між pB і pA віддалена від 0                      | 0.05 / 0.01 | Дає напрям ефекту і його значущість           |
| **t-test (Welch)**     | Числова       | Середні значення          | Чи відрізняються середні           | Порівнює різницю середніх із шумом усередині груп                                                     | 0.05        | Середні статистично різні / ні                |
| **Mann–Whitney U**     | Числова       | Асиметричні розподіли     | Чи одна група “вища” за іншу       | Порівнює **ранги**, а не значення                                                                     | 0.05        | Один розподіл стохастично більший             |
| **Bootstrap**          | Будь-яка      | Складні розподіли         | Який можливий діапазон ефекту      | Багато разів “перевинаходить” вибірку і дивиться, де лежить ефект                                     | (через CI)  | Дає інтервал можливих ефектів                 |


```python
import pandas as pd

df = pd.read_csv("/Users/alksandr/Desktop/kdt_lessons/ab_test.csv")
```

```python
df.columns = ["user_id", "timestamp", "group", "landing_page", "converted"] # rename
df.head()
```

```python
table = pd.crosstab(df["group"], df["converted"])
table
```

```python
# X^2 test тільки для бінарних данних
from scipy.stats import chi2_contingency

chi2, p_value, dof, expected = chi2_contingency(table)
chi2, p_value

# p_value < α -> незалежність відхиляємо -> є ефект
# p_value ≥ α -> недостатньо доказів
```

```python
# Fisher test - для маленьких A/B
from scipy.stats import fisher_exact

odds_ratio, p_value = fisher_exact(table)
odds_ratio, p_value

# p_value < α -> значуща різниця
# odds_ratio > 1 -> B краще за A
# odds_ratio < 1 ->  B гірше
```

```python
# Z-test
import numpy as np
from scipy.stats import norm


conv_A = df[df.group=="control"]["converted"]
conv_B = df[df.group=="treatment"]["converted"]

p1 = conv_A.mean()
p2 = conv_B.mean()

n1 = len(conv_A)
n2 = len(conv_B)

p_pool = (conv_A.sum() + conv_B.sum()) / (n1 + n2)

se = np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))
z = (p2 - p1) / se

p_value = 1 - norm.cdf(z) 
z, p_value
```

```python
# T-test  — для середніх значень, метрика числова (revenue, time, score), приблизна нормальність або велика вибірка
# # !!Приклад коду, не запускати!!
from scipy.stats import ttest_ind

rev_A = df[df.group=="A"]["revenue"]
rev_B = df[df.group=="B"]["revenue"]

t_stat, p_value = ttest_ind(
    rev_B, rev_A,
    equal_var=False  
)

t_stat, p_value
```

```python
# Mann–Whitney U, сильна асиметрія, багато нулів, довгий хвіст 
# !!Приклад коду, не запускати!!
from scipy.stats import mannwhitneyu

rev_A = df[df.group=="A"]["revenue"]
rev_B = df[df.group=="B"]["revenue"]

u_stat, p_value = mannwhitneyu(
    rev_B, rev_A,
    alternative="greater"
)

u_stat, p_value
```

```python
# Bootstrap
import numpy as np

def bootstrap_diff(a, b, n_boot=10000):
    diffs = []
    for _ in range(n_boot):
        a_sample = np.random.choice(a, size=len(a), replace=True)
        b_sample = np.random.choice(b, size=len(b), replace=True)
        diffs.append(b_sample.mean() - a_sample.mean())
    return np.percentile(diffs, [2.5, 97.5])

ci = bootstrap_diff(conv_A.values, conv_B.values)
ci

# CI не містить 0 -> ефект є
# CI містить 0 -> незначимо
```
