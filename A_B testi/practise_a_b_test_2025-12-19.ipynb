#### z-test (–¥–ª—è –ö–û–ù–í–ï–†–°–Ü–ô, 0/1)

```
p = x / n
```

---

#### –†—ñ–∑–Ω–∏—Ü—è –∫–æ–Ω–≤–µ—Ä—Å—ñ–π

```
delta = p_B - p_A
```

---

#### –û–±‚Äô—î–¥–Ω–∞–Ω–∞ (pooled) –∫–æ–Ω–≤–µ—Ä—Å—ñ—è

```
p_pooled = (x_A + x_B) / (n_A + n_B)
```

---

#### –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∞ –ø–æ—Ö–∏–±–∫–∞ (SE)

```
SE = sqrt( p_pooled * (1 - p_pooled) * (1/n_A + 1/n_B) )
```

---

#### z-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞

```
z = delta / SE
```

---

#### p-value (–¥–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω—ñ–π —Ç–µ—Å—Ç)

```
p_value = 2 * (1 - NormalCDF( |z| ))
```

### 95% –¥–æ–≤—ñ—Ä—á–∏–π —ñ–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è —Ä—ñ–∑–Ω–∏—Ü—ñ –∫–æ–Ω–≤–µ—Ä—Å—ñ–π
```
SE_diff = sqrt(
    p_A * (1 - p_A) / n_A +
    p_B * (1 - p_B) / n_B
)

CI = delta ¬± 1.96 * SE_diff
```

---

#### t-test (–¥–ª—è –ß–ò–°–õ–û–í–ò–• –ú–ï–¢–†–ò–ö)

#### –†—ñ–∑–Ω–∏—Ü—è —Å–µ—Ä–µ–¥–Ω—ñ—Ö

```
delta = mean_B - mean_A
```

---

#### –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∞ –ø–æ—Ö–∏–±–∫–∞ (Welch t-test)

```
SE = sqrt( (std_A^2 / n_A) + (std_B^2 / n_B) )
```

---

#### t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞

```
t = delta / SE
```

---

````python
# Task 
# –ü–æ—Ä–∞—Ö—É–≤–∞—Ç–∏ —Å–µ—Ä–µ–¥–Ω—î —É –≥—Ä—É–ø—ñ A —ñ B
# –ó–∞—Å—Ç–æ—Å—É–≤–∞—Ç–∏ Welch t-test
# –ü–æ–¥–∏–≤–∏—Ç–∏—Å—è –Ω–∞ p-value
# –ó—Ä–æ–±–∏—Ç–∏ –≤–∏—Å–Ω–æ–≤–æ–∫

import numpy as np
from scipy import stats

# –î–∞–Ω—ñ
A = np.array([10, 12, 11, 9, 10])
B = np.array([13, 14, 12, 15, 14])

# –ö—Ä–æ–∫ 1: —Å–µ—Ä–µ–¥–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è
mean_A = np.mean(A) # use mean
mean_B = np.mean(B) # use mean

print("–°–µ—Ä–µ–¥–Ω—ñ–π —á–µ–∫ A:", mean_A)
print("–°–µ—Ä–µ–¥–Ω—ñ–π —á–µ–∫ B:", mean_B)

# –ö—Ä–æ–∫ 2: t-test
t_stat, p_value = stats.ttest_ind(A, B, equal_var=False)
print("t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:", t_stat)
print("p-–∑–Ω–∞—á–µ–Ω–Ω—è:", p_value)

````

–≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —ç—Ç–æ —Å–ª—É—á–∞–π–Ω–æ—Å—Ç—å ‚Äî ~0.2%

–ú—ã –ø–æ—á—Ç–∏ –Ω–∞–≤–µ—Ä–Ω—è–∫–∞ –º–æ–∂–µ–º —Å–∫–∞–∑–∞—Ç—å, —á—Ç–æ –≤–∞—Ä–∏–∞–Ω—Ç B –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫, –∞ –Ω–µ –ø—Ä–æ—Å—Ç–æ ¬´–ø–æ–≤–µ–∑–ª–æ —Å –≤—ã–±–æ—Ä–∫–æ–π¬ª.

````python
# Task 

# –ö—Ä–æ–∫–∏
# 1. –ü–æ—Ä–∞—Ö—É–≤–∞—Ç–∏ —Å–µ—Ä–µ–¥–Ω—ñ–π —á–∞—Å
# 2. –ó–∞—Å—Ç–æ—Å—É–≤–∞—Ç–∏ Welch t-test
# 3. –û—Ü—ñ–Ω–∏—Ç–∏ p-value
# 4. –Ü–Ω—Ç–µ—Ä–ø—Ä–µ—Ç—É–≤–∞—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç

A = np.array([40, 42, 38, 41, 39, 40])
B = np.array([43, 44, 45, 41, 42, 43])
print("–°–µ—Ä–µ–¥–Ω—î A:", mean_A)
print("–°–µ—Ä–µ–¥–Ω—î B:", mean_B)
# –ö—Ä–æ–∫ 1. –ü–æ—Ä–∞—Ö—É–≤–∞—Ç–∏ mean, use mean
mean_A = np.mean(A) # use mean
mean_B = np.mean(B)


# –ö—Ä–æ–∫ 2. –ó—Ä–æ–±–∏—Ç–∏ —Ç–µ—Å—Ç, use stats.ttest_ind(equal_var=False)

t_stat, p_value = stats.ttest_ind(A, B, equal_var=False)

print("t_stat =", t_stat)
print("p_value =", p_value)
````

1. p-–∑–Ω–∞—á–µ–Ω–∏–µ 0.00429 < 0.05

(–≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Å–ª—É—á–∞–π–Ω–æ—Å—Ç–∏ ‚âà 0.4%)

2. t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
–ú–∏–Ω—É—Å ‚áí –ø–µ—Ä–≤—ã–π –≤–∞—Ä–∏–∞–Ω—Ç —Ö—É–∂–µ, —á–µ–º –≤—Ç–æ—Ä–æ–π (A < B)
|t| = 3.67 ‚áí —ç—Ñ—Ñ–µ–∫—Ç –∑–∞–º–µ—Ç–Ω—ã–π, –Ω–µ —à—É–º

–í—ã–≤–æ–¥
üìå –ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è
üìå –í–∞—Ä–∏–∞–Ω—Ç B —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–æ –ª—É—á—à–µ –≤–∞—Ä–∏–∞–Ω—Ç–∞ A

–ö–æ—Ä–æ—Ç–∫–æ ¬´–ø–æ-—á–µ–ª–æ–≤–µ—á–µ—Å–∫–∏¬ª
–ú—ã —É–≤–µ—Ä–µ–Ω—ã, —á—Ç–æ —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –≤–∞—Ä–∏–∞–Ω—Ç–∞–º–∏ —Ä–µ–∞–ª—å–Ω–∞—è, –∞ –Ω–µ —Å–ª—É—á–∞–π–Ω–∞—è. –í–∞—Ä–∏–∞–Ω—Ç B –≤—ã–∏–≥—Ä—ã–≤–∞–µ—Ç.

````python
# Task 

# 1. –ü–æ—Ä–∞—Ö—É–≤–∞—Ç–∏ –∫–æ–Ω–≤–µ—Ä—Å—ñ—é
# 2. –ó–∞—Å—Ç–æ—Å—É–≤–∞—Ç–∏ z-test –¥–ª—è –ø—Ä–æ–ø–æ—Ä—Ü—ñ–π
# 3. –ü–æ–¥–∏–≤–∏—Ç–∏—Å—è –Ω–∞ p-value
# 4. –ó—Ä–æ–±–∏—Ç–∏ –≤–∏—Å–Ω–æ–≤–æ–∫

import math
from scipy.stats import norm

nA = 100
xA = 10

nB = 100
xB = 18

# –ö—Ä–æ–∫ 1. –ü–æ—Ä–∞—Ö—É–≤–∞—Ç–∏ –∫–æ–Ω–≤–µ—Ä—Å—ñ—é 
pA = xA / nA # you code 
pB = xB / nB # you code 
print("–ö–æ–Ω–≤–µ—Ä—Å—ñ—è A:", pA)
print("–ö–æ–Ω–≤–µ—Ä—Å—ñ—è B:", pB)

# –ö—Ä–æ–∫ 2: z-test
p_pool = (xA + xB) / (nA + nB)  # you code
SE = math.sqrt(p_pool * (1 - p_pool) * (1/nA + 1/nB))
z = (pA - pB) / SE  # you code
p_value = 2 * (1 - norm.cdf(abs(z)))
print("z-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:", z)
print("p-–∑–Ω–∞—á–µ–Ω–Ω—è:", p_value)
````

üìå –ù–µ–ª—å–∑—è –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É
üìå –ú—ã –Ω–µ –º–æ–∂–µ–º —É–≤–µ—Ä–µ–Ω–Ω–æ —Å–∫–∞–∑–∞—Ç—å, —á—Ç–æ B –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –ª—É—á—à–µ A –ø–æ –∫–æ–Ω–≤–µ—Ä—Å–∏–∏
üìå –ù–∞–±–ª—é–¥–∞–µ–º–∞—è —Ä–∞–∑–Ω–∏—Ü–∞ –º–æ–∂–µ—Ç –±—ã—Ç—å —Å–ª—É—á–∞–π–Ω–æ–π

5. –ü—Ä–æ—Å—Ç—ã–º–∏ —Å–ª–æ–≤–∞–º–∏
–î–∞, —É B –∫–æ–Ω–≤–µ—Ä—Å–∏—è –≤—ã—à–µ (18% vs 10%), –Ω–æ –¥–∞–Ω–Ω—ã—Ö –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ, —á—Ç–æ–±—ã –±—ã—Ç—å —É–≤–µ—Ä–µ–Ω–Ω—ã–º–∏, —á—Ç–æ —ç—Ç–æ –Ω–µ –ø—Ä–æ—Å—Ç–æ —Ñ–ª—É–∫—Ç—É–∞—Ü–∏—è.

````python
# Task 1. –í–∏ —Ç–µ—Å—Ç—É—î—Ç–µ –Ω–æ–≤–∏–π —Ç–µ–∫—Å—Ç –Ω–∞ –∫–Ω–æ–ø—Ü—ñ –æ–ø–ª–∞—Ç–∏.
# –ú–µ—Ç—Ä–∏–∫–∞: purchase conversion (–∫—É–ø–∏–≤ / –Ω–µ –∫—É–ø–∏–≤) ‚Äî –±—ñ–Ω–∞—Ä–Ω–∞.

# –î–∞–Ω—ñ –∑–∞ —á–∞—Å —Ç–µ—Å—Ç—É:
# A (–∫–æ–Ω—Ç—Ä–æ–ª—å): 8 000 –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á—ñ–≤, 720 –æ–ø–ª–∞—Ç
# B (—Ç–µ—Å—Ç): 8 000 –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á—ñ–≤, 784 –æ–ø–ª–∞—Ç–∏

# –ö—Ä–æ–∫ 0. –§–æ—Ä–º—É–ª—é—î–º–æ –≥—ñ–ø–æ—Ç–µ–∑–∏

#  —Ä—ñ–∑–Ω–∏—Ü—ñ –Ω–µ–º–∞—î
#  —Ä—ñ–∑–Ω–∏—Ü—è —î

# –î–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω—ñ–π —Ç–µ—Å—Ç, –±–æ –∑–º—ñ–Ω–∞ –º–æ–∂–µ —è–∫ –ø–æ–∫—Ä–∞—â–∏—Ç–∏, —Ç–∞–∫ —ñ –ø–æ–≥—ñ—Ä—à–∏—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç.
````

````python
# –ö—Ä–æ–∫ 1. –†–∞—Ö—É—î–º–æ –∫–æ–Ω–≤–µ—Ä—Å—ñ—ó —Ç–∞ –¥–µ–ª—å—Ç–∞(pB‚Äã‚àípA)
# –ö—Ä–æ–∫ 2. –û–±‚Äô—î–¥–Ω–∞–Ω–∞ –∫–æ–Ω–≤–µ—Ä—Å—ñ—è (p = (pa + pb) / count A + count B)
# –ö—Ä–æ–∫ 3. –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∞ –ø–æ—Ö–∏–±–∫–∞(SE)
# –ö—Ä–æ–∫ 4. z-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –¥–µ–ª—å—Ç–∞ / SE
# –ö—Ä–æ–∫ 5. p-value (–¥–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω—ñ–π)
````

````python

import numpy as np
from scipy import stats

nA, xA = 8000, 720
nB, xB = 8000, 784

pA = xA / nA
pB = xB / nB
print("–ö–æ–Ω–≤–µ—Ä—Å—ñ—è A:", pA)
print("–ö–æ–Ω–≤–µ—Ä—Å—ñ—è B:", pB)

delta = pB - pA

p_pool = (xA + xB) / (nA + nB)
se = np.sqrt(p_pool * (1 - p_pool) * (1/nA + 1/nB))

z = delta / se
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print("z-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:", z)
print("p-–∑–Ω–∞—á–µ–Ω–Ω—è:", p_value)
# –í–∏—Å–Ω–æ–≤–æ–∫: —è–∫—â–æ p-value < 0.05, –≤—ñ–¥—Ö–∏–ª—è—î–º–æ H0, —ñ–Ω–∞–∫—à–µ –Ω–µ –≤—ñ–¥—Ö–∏–ª—è—î–º–æ H0.
````
