```python
arr = [
    [1, 2, 3]
    , [4, [5, 11, 12], 6]
    , [7, 8, 9]
]
print(arr[1][1][1])
```

```python
arr = [i for i in range(1, 15)]

dp = [0] * len(arr)

dp[0] = arr[0]

for i in range(1, len(arr)):
    dp[i] = dp[i - 1] + arr[i]

print(arr) 
print(dp)
```

```python
# Zeroes matrix
import numpy as np

A = np.zeros((3, 4))
print(A)
```

```python
# Ones matrix
I = np.eye(4)
print(I)
```

```python
# Diag matrix
D = np.diag([3, 7, -2])
print(D)
```

```python
# Upper triangle matrix
U = np.triu([[2, 5, 3],
             [4, 4, 1],
             [8, 2, 7]])
print(U)
```

```python
# Symmetric matrix
A = np.array([[4, 2, 7],
              [2, 5, 1],
              [7, 1, 3]])

print(A)
print(np.allclose(A, A.T))
```

```python
from scipy.sparse import csr_matrix

R = csr_matrix([[5, 3, 0, 0],
                [4, 0, 0, 2],
                [1, 1, 0, 5],
                [0, 0, 5, 4]])

print(R)
print("–§–æ—Ä–º–∞—Ç:", type(R))
```

```python
import numpy as np
# –º–Ω–æ–∂–µ–Ω–Ω—è
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

C = A @ B   # –∞–±–æ np.dot(A, B)
print(C)
# [
#   5*0.3 + 10*0.7 = 8.5
#   2*0.3 +  4*0.7 = 3.4
# ]
```

```python
# rank 
A = np.array([[1, 2, 3],
              [2, 4, 6],
              [3, 6, 9]])

rank = np.linalg.matrix_rank(A)
print("–†–∞–Ω–≥:", rank)
```

```python
X = df[["age", "fare", "pclass"]].values  
y = df["survived"].values                 
```
