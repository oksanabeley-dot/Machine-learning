```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

```python
tips = sns.load_dataset("tips")
sns.histplot(tips["total_bill"], bins=20)
plt.show()
```

```python
sns.boxplot?
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, palette = 'magma')
plt.show();
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, palette = 'magma')
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="sex",
    data=tips
)
plt.show()
```

```python
sns.regplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
sns.countplot(x="day", data=tips, palette = 'magma')
plt.show()
```
