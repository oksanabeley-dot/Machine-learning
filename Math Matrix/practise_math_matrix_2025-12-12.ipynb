Детерминант только для квадратных матриц Главная причина в том, что детерминант можно вычислить только для квадратной матрицы. 
Наша исходная матрица X имеет размерность, например, 891 строка (пассажиры) на 3 столбца (признаки). Она не квадратная.
Матрица же \(X^{T}\cdot X\) всегда является квадратной матрицей. 
Если в \(X\) 3 столбца, то \(X^{T}\cdot X\) будет матрицей 3х3 (по одному измерению для каждого признака). 

Цель проверки: Обратимость матрицы и поиск решения Основная причина проверки детерминанта в этом контексте — это определение обратимости матрицы \(X^{T}\cdot X\). Это критически важно для решения систем линейных уравнений или обучения моделей. 

В линейной регрессии, например, коэффициенты модели \(\beta \) вычисляются по формуле: \(\beta =(X^{T}\cdot X)^{-1}\cdot X^{T}\cdot y\)Чтобы решить эту формулу, нам нужно вычислить обратную матрицу к \(X^{T}\cdot X\) (часть \((X^{T}\cdot X)^{-1}\)). Матрица имеет обратную только в том случае, если ее детерминант не равен нулю. Если детерминант \(X^{T}\cdot X\) близок к нулю (или равен ему), то матрица \(X^{T}\cdot X\) является вырожденной (singular), и ее нельзя обратить математически.Это означает, что мы не можем надежно вычислить коэффициенты модели классическим способом. 

Причина проверки ранга Ранг мы проверяем по той же причине. Если ранг матрицы \(X^{T}\cdot X\) меньше количества столбцов (то есть меньше 3 в вашем случае), это сигнализирует о проблеме мультиколлинеарности (один признак идеально предсказывается другим, например, вы бы добавили и age_in_years, и age_in_months).
